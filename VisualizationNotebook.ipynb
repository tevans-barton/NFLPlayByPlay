```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
nfl_data = pd.read_csv('cleaned_nfl_data.csv')
```

```python
nfl_data = nfl_data.drop('Unnamed: 0', axis = 1)
```

```python
temp = nfl_data[['GameID', 'HomeScore', 'AwayScore', 'HomeTeam', 'AwayTeam']].groupby('GameID').aggregate(max).reset_index()
temp
```

```python
winners = {}
for x in range(len(temp)):
    if temp['HomeScore'][x] > temp['AwayScore'][x]:
        winners[temp['GameID'][x]] = temp['HomeTeam'][x]
    elif temp['HomeScore'][x] < temp['AwayScore'][x]:
        winners[temp['GameID'][x]] = temp['AwayTeam'][x]
    else:
        winners[temp['GameID'][x]] = 'TIE'
```

```python
nfl_data['Winner'] = [winners[nfl_data['GameID'][x]] for x in range(len(nfl_data))]
nfl_data
```

```python
nfl_data.to_csv('cleaned_nfl_data(v2).csv')
```
